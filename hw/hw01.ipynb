## Section 1: Introduction to Tensors

At its core, the tensor in Pytorch is just like NumPy's `ndarray` (n-dimensional array), but with a couple of critical differences for neural nets:
1. Tensors can run on GPUs.
2. Tensors are optimized for automatic differentiation for building and training neural networks.

Let's dive a bit deeper into tensors.

A **tensor** is a specialized data structure for representing n-dimensional data. Yes, that means vectors for 1-d, matrices for 2-d, and so on. In PyTorch, we use **tensors** to encode all of the parameters of the model (i.e. the weights and biases), as well as the input and output of a model. All data you'll be working with must be encoded in tensors first! *Tensors are the most fundamental data structure in PyTorch, and you will use them in every asssignment. #### Indexing and Slicing

PyTorch tensors follow standard NumPy-like indexing and slicing.


````python
# Initialize the tensor with values 1, 2, 3, etc. not ones.
tensor = torch.arange(1,21).reshape([5,4])
print(tensor)
print('First row: ', tensor[0])
print('First column: ', tensor[:, 0])
print('First column: ', tensor[..., 0])
print('Last column:', tensor[..., -1])
tensor[:, 1] = 0
print(tensor)
````

#### Joining tensors

You can use `torch.cat` to concatenate a sequence of
tensors along a given dimension. See also
[torch.stack](https://pytorch.org/docs/stable/generated/torch.stack.html),
another tensor joining operator that is subtly different from
`torch.cat`.


````python
tensor = torch.arange(1,13).reshape(2,6)
print("Original tensor:\n", tensor)
t1 = torch.cat([tensor, tensor, tensor], dim=-1)
print(t1)

t4 = torch.cat([tensor, tensor, tensor], dim=0)
print(t4)
# Demonstrating torch.stack as well
# torch.cat concatenates tensors along an existing dimension.
# torch.stack joins a sequence of tensors along a new dimension.

# Let's stack three copies of the tensor along dim=0
t2 = torch.stack([tensor, tensor, tensor], dim=0)
print("\nStacked tensor shape:", t2.shape)
print("Stacked tensor (dim=0):\n", t2)

# You can also stack along another dimension, e.g., last dimension
t3 = torch.stack([tensor, tensor, tensor], dim=-1)
print("\nStacked tensor shape (dim=-1):", t3.shape)
print("Stacked tensor (dim=-1):\n", t3)

````

#### Matrix Multiplication Reference Table

| Multiplication Name | PyTorch Function / Operator | Description |
| :--- | :--- | :--- |
| **Element-wise Multiplication** | `*` or `torch.mul` | Multiplies corresponding elements of two tensors of the same shape (Hadamard product). |
| **Dot Product** | `torch.dot` | Computes the inner product of two 1D tensors (vectors). |
| **Matrix Multiplication** | `@` or `torch.matmul` | Performs standard matrix-matrix multiplication, including broadcasting for batches. |
| **Batch Matrix Multiplication** | `torch.bmm` | Performs matrix multiplication on two batches of matrices without broadcasting. |
| **Matrix-Vector Product** | `torch.mv` | Performs a multiplication of a matrix by a vector. |
| **Outer Product** | `torch.outer` | Computes the outer product of two vectors, resulting in a matrix. |

**Key Concepts:**
*   **Batch**: A collection of multiple samples (like images or data rows) grouped together in a single tensor. In PyTorch, the first dimension is typically the "batch dimension" (e.g., a shape of `[64, 3, 224, 224]` represents a batch of 64 color images).
*   **Matrix Multiplication Rules**: To multiply matrix $A$ (shape $m \times n$) by matrix $B$ (shape $p \times q$), the "inner" dimensions must match ($n = p$). The resulting matrix will have the "outer" dimensions ($m \times q$).
*   **Inner vs. Outer Product**:
    *   **Inner Product (Dot Product)**: Takes two vectors and returns a single scalar (sum of element-wise products).
    *   **Outer Product**: Takes two vectors of size $m$ and $n$ and returns an $m \times n$ matrix where each element $(i, j)$ is the product of the $i$-th element of the first vector and the $j$-th element of the second.
*   **Broadcasting**: A mechanism that allows PyTorch to perform operations on tensors of different shapes by automatically "expanding" the smaller tensor to match the dimensions of the larger one without copying data.


````python
# For matrix multiplication, let's prepare two tensors
v = torch.tensor([5, 10, 15])
a = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])  # shape (2×3)
b = torch.tensor([[7, 8],
                  [9, 10],
                  [11, 12]])  # shape (3×2)

# Demonstrate an inner product (dot product) with v
v_inner = torch.dot(v, v)
print("\nInner product of v with itself:", v_inner)

# Demonstrate a standard multiplication on v
v_mul = v * v
print("\nElement-wise multiplication of v by itself:", v_mul)

# Matrix multiplication: result shape is (2×2)
mat_mult = a @ b
print("Matrix multiplication with the @ operator:\n", mat_mult)

# Or use the torch.matmul function
print("Matrix multiplication using torch.matmul:\n", torch.matmul(a, b))

# A few more examples, using the transform with .T
print("A matrix multiplied by its transpose:")
print(a @ a.T)

# Demonstrate a matrix vector product, which is nothign more than a basic matrix multiplication
# but one matrix is a vector
print("Matrix vector product:")
print(a.mv(v))
````

**In-place operations** Operations that store the result into the
operand are called in-place. They are denoted by a `_` suffix. For
example: `x.copy_(y)`, `x.t_()`, will change `x`.


````python
print("Original tensor a:\n", a, "\n")

# Standard addition (creates a new tensor, does not modify 'a')
a_added = a + 5
print("Result of a + 5 (not inplace):\n", a_added)
print("Tensor a after standard addition (should be unchanged):\n", a, "\n")

# In-place addition (modifies 'a' directly)
a.add_(5)
print("Tensor a after in-place addition (a.add_(5)):\n", a)
````


**What is a batch?**<br>
The term "batch" is basically a collection of multiple matrices processed together. Thus, batches are usually tensors with more than two dimensions. This is oversimplified, but it's a good starting point.

Think about input data processing. You might have millions of examples that need to be fed into your neural network, each example has perhaps 1000 features. You can't feed the entire dataset all at once, so you batch them. It's common to have batches of 32, 64, 128, 256, 512, 1024, etc. examples. The first dimension of the tensor is the batch dimension. This is particularly useful for training neural networks on GPUs, where you can process multiple examples in parallel. It's also useful when dealing with color images, where each pixel is a feature over 3 values (red, green, blue). So a batch of 32 images is a 32x32x3 tensor.


For example, if you have a batch of 5 matrices, each 4x4, you can create a tensor with shape (5, 4, 4).
```python
batch = torch.randn(5, 4, 4)
```
This creates a 3D tensor with shape (5, 4, 4), which means:
* 5 matrices (the batch dimension)
* Each matrix is 4×4<br>

So you have 5 4x4 matrices stacked together. The first dimension (5) is the batch dimension.

**Why use batches?**<br>
* Generally, it comes down to efficiency: Process multiple examples at once using parallel operations. It's quite common in neural networks, especially training ones. Neural networks typically process data in batches rather than one example at a time
* Vectorization: PyTorch can perform operations on all matrices in the batch simultaneously

**What is broadcasting?**<br>

Broadcasting is a technique used by PyTorch (and other array libraries like NumPy) to perform operations on tensors of different shapes. When operating on two tensors, PyTorch will automatically expand the smaller tensor's shape so that it matches the larger tensor's shape in a compatible way, without actually copying data. This is very useful for element-wise operations like addition, subtraction, etc.

**Rules of Broadcasting:**
- Starting from the trailing dimensions, two dimensions are compatible when they are equal, or one of them is 1.
- If necessary, the dimensions of size 1 are stretched or "broadcast" to match the other tensor.


**Example:**<br>
Create a 3x4 tensor filled with the value 2 named 'a' and a 1D tensor of size 4 with values [1, 2, 3, 4] named 'b'. Add the two tensors together. Print the result and explain the shape of the output.


````python
a = torch.full((3, 4), 2)
print("Tensor a:\n", a)
#
# Create a 1D tensor of size 4
b = torch.tensor([1, 2, 3, 4])
print("Tensor b:\n", b)
#
# Add using broadcasting
# 1. Starting from the last dimension, b (size 4) matches a (size 4).
# 2. b has no first dimension, so PyTorch treats it as having size 1.
# 3. This size 1 is "broadcast" (virtually repeated) 3 times to match a.
# 4. Effectively, [1, 2, 3, 4] is added to each of the 3 rows in 'a'.
result = a + b
print("Result of a + b:\n", result)
print("Result shape:", result.shape)
````

### Is a GPU always better?

OK, so this is the question you should be asking. And, well, it depends! GPUs (Graphics Processing Units) are designed for highly parallel computations, meaning they can perform many calculations simultaneously. CPUs (Central Processing Units), while very powerful, are optimized for sequential tasks and complex logic.

So, one might think you should always use a GPU, right? It's not that simple.

#### GPUs explained

GPUs are all about having the power to solve computations with **massive parallelism**! GPUs have thousands of smaller, specialized cores, unlike CPUs which have fewer, but more powerful cores. This architecture allows them to process many pieces of data concurrently.

#### Where do GPUs shine?

GPU cores are highly optimized for mathematical operations, especially matrix multiplications and linear algebra computations, including basic matrix additions and multiplications. Functions that can be parallelized shine quite well on GPUs. These operations are fundamental to fields like graphics rendering, scientific simulations, and yes, **neural networks**, especially large, deep nets.

#### Do GPU computations have a price?

There is a big **Data Transfer Overhead**! For the GPU to perform computations, data first needs to be explicitly moved from the CPU's main memory to the GPU's dedicated high-speed memory (VRAM). After computation, the results often need to be transferred back. This data transfer introduces an inherent overhead.

#### When does a GPU Outperform a CPU?

Generally, you should seek GPU support when:

* **Workload is Large and Highly Parallel**: The task involves a huge number of independent, similar calculations that can be performed at the same time. Examples include processing large images, training deep neural networks, or running complex simulations. The larger the dataset and the more independent calculations involved, the more the GPU's parallel architecture shines.

* **Computation Time Significantly Exceeds Data Transfer Overhead**: The time saved by parallel execution on the GPU must be much greater than the time spent moving data to and from the GPU's memory. For smaller tasks, or tasks that require frequent data transfers back and forth between the CPU and GPU, the overhead can negate or even reverse the GPU's speed advantage.

* **High Throughput Required**: When the goal is to process a massive amount of data or perform a large number of computations in the shortest possible time, GPUs excel at achieving high throughput.

In summary, while CPUs are versatile and efficient for a wide range of tasks, GPUs become indispensable for workloads that are (to quote one thread I read), 'embarrassingly parallel' and involve heavy numerical computation on large datasets. **The more intense and parallelizable the computation, the greater the performance gap between a GPU and a CPU will be.**

## Section 3: Automatic Differentiation (Autograd)

`torch.autograd` is PyTorch’s built-in differentiation engine that powers neural network training. It automatically computes gradients for any computational graph. You will soon appreciate the important of highly efficient differentiation!

### What is Automatic Differentiation?

**Automatic differentiation (autodiff)** is a technique for efficiently computing derivatives of functions specified by computer programs. Unlike symbolic differentiation (which manipulates mathematical expressions) or numerical differentiation (which uses finite differences), autodiff computes exact derivatives while maintaining computational efficiency.

In the context of neural networks, we need to compute gradients of a loss function with respect to thousands or even millions of parameters. Doing this by hand would be intractable. Automatic differentiation solves this problem elegantly.

### Why is Autograd Critical for Deep Learning?

Neural network training relies on an algorithm called **backpropagation**, which is essentially the chain rule of calculus applied repeatedly. Here's why autograd is essential:

1. **Gradient-Based Optimization**: Training a neural network means finding parameters (weights and biases) that minimize a loss function. This requires computing gradients of the loss with respect to every parameter.

2. **Complex Computational Graphs**: Modern neural networks involve complex operations—matrix multiplications, non-linear activations, convolutions, attention mechanisms, etc. Computing gradients through these operations manually is error-prone and tedious.

3. **Efficiency**: Autograd computes gradients in a single backward pass through the computational graph, making it highly efficient even for very deep networks.

4. **Flexibility**: PyTorch's dynamic computational graph (define-by-run) means you can change your network architecture on the fly, and autograd will still work correctly.


### How Does PyTorch's Autograd Work?

PyTorch uses a technique called **reverse-mode automatic differentiation**. Here's the key idea:

#### 1. Building the Computational Graph

When you perform operations on tensors with `requires_grad=True`, PyTorch builds a **directed acyclic graph (DAG)** called the computational graph:
- **Nodes** represent tensors (data)
- **Edges** represent operations (functions)

Each operation creates a `grad_fn` object that knows how to compute gradients!

````python
import torch
# Example: Building a simple computational graph
x = torch.tensor(3.0, requires_grad=True)
y = torch.tensor(4.0, requires_grad=True)

# Operations build the graph
z = x * y + y ** 2
print(f"z = {z}")
print(f"z.grad_fn = {z.grad_fn}")  # Shows the operation that created z
````

In this example:
- `x` and `y` are **leaf tensors** (created by the user, not by operations)
- `z` is a **non-leaf tensor** created by operations
- `z.grad_fn` is `<AddBackward0>`, representing the addition operation

#### 2. The Backward Pass

When you call `.backward()` on a scalar tensor (like a loss), PyTorch:
1. Starts at that tensor
2. Traverses the graph backward (from outputs to inputs)
3. Applies the chain rule at each node to compute gradients
4. Accumulates gradients in the `.grad` attribute of each leaf tensor

This is going to make so much more sense when we work through backpropagation - is the process by which neural nets computes gradients for all tensors in the computational graph, starting from the loss and moving backward through the operations that created them. This allows us to efficiently compute gradients for complex models with many interconnected operations.

Think about this mathematically for a second. Our basic function, z, is:

$$z = x \cdot y + y^2$$

The derivative of z with respect to x is y, and with respect to y is x + 2y. Let's verify this, but first we need to explain the `.backward()` method:

#### Understanding `z.backward()`

When we call `z.backward()`, PyTorch performs the **backward pass** through the computational graph:

1. **Starting Point**: It begins at tensor `z` (our output) and works backward toward the inputs `x` and `y`.

2. **Chain Rule Application**: At each operation node in the graph, PyTorch applies the chain rule to compute how much each input contributed to the final value. For our function $z = x \cdot y + y^2$:
   - The gradient $\frac{\partial z}{\partial x} = y$ (derivative of $xy$ with respect to $x$)
   - The gradient $\frac{\partial z}{\partial y} = x + 2y$ (sum of derivatives: $\frac{\partial(xy)}{\partial y} + \frac{\partial(y^2)}{\partial y}$)

3. **Gradient Storage**: The computed gradients are stored in the `.grad` attribute of each **leaf tensor** (tensors created by the user with `requires_grad=True`). In our case, `x.grad` and `y.grad` will contain the respective partial derivatives.

4. **Graph Destruction**: After `.backward()` completes, the computational graph is destroyed by default to free memory (unless `retain_graph=True` is specified).

This automatic differentiation is what makes training neural networks feasible—PyTorch handles all the calculus for us!

OK, let's verify this... and be amazed at the simplicity of how Pytorch carries all of this out for us:

````python
# Continuing from above

# z.backward 
z.backward()

# Let's verify mathematically:
# z = x*y + y^2
# dz/dx = y = 4.0
# dz/dy = x + 2y = 3 + 2*4 = 11.0

print(f"dz/dx = {x.grad}")  # Should be y = 4.0
print(f"dz/dy = {y.grad}")  # Should be x + 2y = 3 + 2*4 = 11.0

# Source: https://medium.com/@namanphy/understanding-autograd-5-pytorch-tensor-functions-8f47c27dc38
````

### Key Concepts in PyTorch's Autograd

1. **Leaf Tensors**: Tensors created directly by the user. Only leaf tensors with `requires_grad=True` will have gradients stored in `.grad`.

2. **Non-Leaf Tensors**: Tensors created by operations. Their gradients are not stored by default (to save memory), but they have `grad_fn` to track how they were created.

3. **Gradient Accumulation**: Calling `.backward()` multiple times will **accumulate** gradients in `.grad`. You need to call `.zero_grad()` to clear them.

````python
# Demonstrating gradient accumulation
x = torch.tensor(2.0, requires_grad=True)

for i in range(3):
    y = x ** 2 + 2 * x
    # dy/dx = 2x + 2
    y.backward()
    print(f"Iteration {i + 1}: x.grad = {x.grad}")

# Notice gradients accumulate! To prevent this:
x.grad.zero_()
print(f"After zeroing: x.grad = {x.grad}")
````

4. **Detaching from the Graph**: Sometimes you want to use a tensor's value without tracking gradients. Use `.detach()` to create a new tensor that shares data but has no gradient tracking.

````python
x = torch.tensor(3.0, requires_grad=True)
y = x ** 2

# Detach y from the computational graph
y_detached = y.detach()
print(f"y_detached requires_grad: {y_detached.requires_grad}")

z = y_detached * 2

# Now z has no connection to x, so gradients won't flow back
# z.backward() would fail because z is not connected to any requires_grad=True tensor
````

### Memory Management

PyTorch automatically manages the computational graph:
- The graph is built dynamically during the forward pass
- After `.backward()`, the graph is destroyed by default (to save memory)
- For multiple backward passes, use `retain_graph=True` in `.backward()`

### Practical Implications

Understanding autograd is crucial because:
1. **Debugging**: You need to understand when gradients flow and when they don't
2. **Custom Operations**: You may need to define custom backward passes for novel operations
3. **Optimization**: Knowing when to detach or use `torch.no_grad()` can save memory and computation
4. **Advanced Architectures**: Some architectures (like GANs) require careful gradient management

Don't get too concerned about the details right now. Just know that the automatic differential is absolutely critical for neural network training.
